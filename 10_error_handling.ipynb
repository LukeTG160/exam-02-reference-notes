### try / except (no specific exception type)

```python
try:
    x = int("123")  # change to "abc" to force an error
    print(x)
except:
    print("Something went wrong")
```

### Loop until valid input (try / except in a loop)

```python
while True:
    try:
        qty = int(input("Enter quantity (integer): "))
        print("Quantity accepted:", qty)
        break
    except:
        print("Invalid input. Please enter a whole number.")
```

### Defensive checks vs try/except

```python
# You can use a defensive approach or try/except exception handling approach on any exam.

# Defensive approach
user_text = "42"
if user_text.isdecimal():
    print(int(user_text))
else:
    print("Not a whole number")

# try/except approach
user_text2 = "42x"
try:
    print(int(user_text2))
except ValueError:
    print("Conversion failed (ValueError)")
```

### Common exception types (ValueError, KeyError, ZeroDivisionError, TypeError)

```python
# ValueError
try:
    int("abc") # improper text value for an integer
except ValueError as e:
    print("ValueError:", e)

# KeyError
data = {"sku":"A100", "price":29.0}
try:
    print(data["stock"])  # key doesn't exist
except KeyError as e:
    print("KeyError:", e)

# ZeroDivisionError
try:
    10 / 0 # dividing by zero isn't allowed
except ZeroDivisionError as e:
    print("ZeroDivisionError:", e)

# TypeError
try:
    "10" + 5 # incompatible operation for 2 different data types
except TypeError as e:
    print("TypeError:", e)
```

### Catch any exception and inspect the exception object

```python
try:
    open("definitely_not_a_file.txt")
except Exception as e:
    print(type(e).__name__, "-", e)
```
