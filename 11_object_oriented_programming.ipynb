### Define a class & create objects

```python
class Product:
    def __init__(self, sku, price): # must be called __init__. Also known as the "constructor" or "initializer"
        self.sku = sku
        self.price = price

p1 = Product("A100", 29.0)
p2 = Product("B200", 45.0)
print(p1.sku, p1.price)
print(p2.sku, p2.price)
```

### Class variables vs instance variables

```python
class Order:
    tax_rate = 0.0725 # this is a class variable
    def __init__(self, subtotal):
        self.subtotal = subtotal # this is an instance variable

o = Order(100)
print(Order.tax_rate)
print(o.tax_rate)
print(o.tax_rate)
```

### Writing Methods

```python
class CartItem:
    def __init__(self, name, price, qty):
        self.name = name
        self.price = price
        self.qty = qty
        
    def total(self):
        return self.price * self.qty

item = CartItem("Candle", 19.99, 3)
print(item.total())
```

### Put objects in a list

```python
class Product:
    def __init__(self, sku, price):
        self.sku = sku
        self.price = price

p1 = Product("A100", 29.0)
p2 = Product("B200", 45.0)

p_list = [p1, p2]
print(p_list[0].sku) # example of accessing things inside of a list

# looping through a list and accessing each object
for obj in p_list:
    print(obj.price)
```

### Put objects in a dictionary

```python
class Product:
    def __init__(self, sku, price):
        self.sku = sku
        self.price = price

p1 = Product("A100", 29.0)
p2 = Product("B200", 45.0)

# the sku is the key, the whole object is the value
products = {p1.sku: p1, p2.sku: p2}

print(products["A100"].price) # access a specific instance value of an object
products["B200"].price = 49.0
print(products["B200"].price)
```
